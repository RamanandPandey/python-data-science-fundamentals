# Quick QNA?

List vs NumPy array?

A List is a built-in data type in Python that can store elements of different data types. It is flexible but slower compared to NumPy arrays because operations are performed using loops.

NumPy is a Python library used for numerical computations on arrays. A NumPy array stores elements of the same data type and performs operations much faster because it uses vectorized operations instead of loops.

    

Why NumPy faster?

NumPy is faster than normal Python lists because The Backend of numpy has written in c.
And also because it stores data in contiguous memory, 
ses a single data type, and performs vectorized operations with optimized low-level loops, 
which together reduce overhead and make computations much faster.

What is DataFrame?

A DataFrame is a two-dimensional data structure in Python that organizes data in the form of rows and columns. 
It is similar to an Excel sheet and is widely used for data analysis tasks such as data cleaning, 
transformation, and performing analytical operations.

.loc vs .iloc?

.loc is a pandas method used for label-based indexing, 
where rows and columns are accessed using their names (labels). 
For example: df.loc[0, 'Marks'].

.iloc is a pandas method used for integer position-based indexing, 
where rows and columns are accessed using their numerical index positions. 
For example: df.iloc[0, 1].
